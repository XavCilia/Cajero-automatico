#### **Cuaderno 3**: El método de la bisección

En muchos problemas matemáticos y físicos es necesario encontrar «las raíces» de una función, es decir, un valor que haga que la función se anule.

Existen diferentes alternativas para encontrar las raíces de una función. Una de ellas (aunque lejos de ser la mejor), es el llamado «__método de la bisección__».

En este cuaderno, veremos en qué consiste y lo programaremos para encontrar la raíz de una función.

Supongamos que tenemos la siguiente función:

$$ f(x) = x^3-3x^2+3x-4 $$

A continuación, vamos a ver la representación gráfica de esta función:

_**Nota**: El código que se muestra a continuación es complejo, pero no le prestaremos atención por el momento._

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = x**3 - 3 * x**2 + 3 * x - 4

plt.rcParams['figure.figsize'] = [6, 4]
plt.rcParams['figure.dpi'] = 120
plt.grid(axis='both', color='0.95')
plt.ylim([-60, 60])
plt.plot(x, x*0, 'k')
plt.plot(x, y, 'r')
plt.show()
```

En la figura anterior, podemos ver que la función (dibujada en rojo), se anula en un punto que se encuentra en algún lugar entre $x = 2$ y $x = 4$. Dicho punto es aquél en el que se corta la función con la recta $y = 0$ (dibujada en negro).

Ahora bien, ¿cómo podemos descubrir cuál es el valor exacto de $x$ para el cual $f(x) = 0$? Lo cierto es que, en este caso, hay procedimientos analíticos para hacerlo. Sin embargo, vamos a emplear el __método de la bisección__.

Este método se basa en el Teorema de Bolzano, que afirma:

_Sea $f\,:\,[a, b] \rightarrow \mathbb{R}$ una función real continua en $[a, b]$ con $f(a) < 0 < f(b)$. Entonces existe al menos un punto $c \in (a, b)$ tal que $f(c) = 0$._

Dicho de forma coloquial, si tenemos una función continua (la de nuestro ejemplo lo es) y conocemos un intervalo para el cuál el valor es negativo al comienzo del intervalo y positivo al final del intervalo, entonces la función debe tomar el valor 0 en algún punto del intervalo.

Esto lo observamos con claridad en el intervalo $[2, 4]$ de la figura anterior.

**¿Cómo funciona el método de la bisección?**

El funcionamiento de este método es muy sencillo: partimos de un intervalo en el que sepamos que existe una raíz y lo hacemos cada vez más estrecho, hasta encontrar un valor que anula la función.

Para ello, realizamos el siguiente procedimiento:
1. Partimos del intervalo $[a, b]$.
2. Calculamos el punto medio del intervalo, que será $m = \frac{a + b}{2}$.
3. Calculamos el valor de $f(m)$.
  + Si $f(m) > 0$, entonces la raíz debe encontrarse en el intervalo $[a, m]$.
  + Si $f(m) < 0$, entonces la raíz debe encontrarse en el intervalo $[m, b]$.

El proceso lo repetimos, sustituyendo el intervalo inicial por el nuevo (que tendrá justo _la mitad del ancho_) que el original.

Al hacer el intervalo cada vez más estrecho, nos aseguramos estar acotando cada vez mejor el valor que anula a la función. En algún momento, encontraremos el valor exacto, o al menos una aproximación muy buena.

A continuación, vas a programar el método de la bisección:

```python

a = 2
b = 4
E = 1E-8 # Este es el margen de error máximo que buscamos.

# Recordemos que la función empleada es: x**3 - 3 * x**2 + 3 * x - 4

# En primer lugar, calcularemos el punto medio –m– del intervalo
#### ESCRIBE EL CÓDIGO AQUÍ (1 línea)
m = (a +b ) / 2

####

# A continuación, calcularemos el valor de la función en dicho punto medio –f(m).
#### ESCRIBE EL CÓDIGO AQUÍ (1 línea)
fm = m**3 - 3 * m**2 + 3 * m - 4
####

# A continuación, iniciaremos un bucle que debe repetirse mientras el error
# sea mayor que E.
# Nota: ¡el error es la diferencia |f(m) - 0| en valor absoluto!
# Puedes calcular el valor absoluto de un número `x` empleando la función `abs(x)`.
#### ESCRIBE EL CÓDIGO AQUÍ (1 línea)
while abs(fm) > E:
####

  # A continuación, determinaremos el nuevo intervalo.
  # Habrá que modificar el valor de a o el valor de b, en función de si
  # f(m) < 0 o f(m) > 0.
  #### ESCRIBE EL CÓDIGO AQUÍ (4 líneas)
  if fm < 0:
    a = m
  else:
    b = m

  ####

  # Por último, calculamos el nuevo valor de m y de f(m).
  # Nota: se deben reemplazar los valores calculados anteriormente.
  #### ESCRIBE EL CÓDIGO AQUÍ (2 líneas)
m = (a +b ) / 2
fm = m**3 - 3 * m**2 + 3 * m - 4
  ####

# Para finalizar, imprimiremos el valor calculado para la raíz:
print(f"La función se anula en el punto x = {m:.4f}")
```

El punto de corte de esta función se encuentra aproximadamente en el punto $x = 2.4422$. Si has encontrado el valor, ¡enhorabuena!

Si no lo has encontrado, no desesperes. Asegúrate de haber calculado bien los valores de la función y los nuevos intervalos en cada iteración.
